# Titanic Survival Analysis

This project analyzes the Titanic dataset to understand passenger survival patterns based on gender, passenger class, and age. Data cleaning and visualization techniques were applied using Python.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df.isnull().sum()
```

```python
df.groupby("Sex")["Survived"].mean()
```

```python
df.groupby("Pclass")["Survived"].mean()
```

```python
df.groupby(pd.cut(df["Age"], bins=[0,18,35,60,100]),observed=False)["Survived"].mean()
```

```python
sns.barplot(x="Sex", y="Survived", data=df)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
sns.barplot(x="Pclass", y="Survived", data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
df["Age"].plot(kind="hist", bins=20)
plt.title("Passenger Age Distribution")
plt.xlabel("Age")
plt.show()
```

# Conclusion

The analysis shows that survival rates varied significantly by gender and passenger class. Female passengers and higher-class passengers had better survival chances. Visualizations helped identify these patterns clearly.
